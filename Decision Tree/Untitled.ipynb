```python
import pandas as pd
df=pd.read_csv('salaries.csv')
df
```

```python
inputs=df.iloc[:,:-1]
inputs
```

```python
output=df.iloc[:,-1]
output
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
company_le=LabelEncoder()
job_le=LabelEncoder()
degree_le=LabelEncoder()
df['new_company']=company_le.fit_transform(df['company'])
df['new_job']=job_le.fit_transform(df['job'])
df['new_degree']=degree_le.fit_transform(df['company'])
```

```python
df
```

```python
df=df.drop(['company','job','degree'],axis='columns')
df
```

```python
new_inputs=df[['new_company','new_job','new_degree']]
new_inputs
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(new_inputs,output,test_size=0.2)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
