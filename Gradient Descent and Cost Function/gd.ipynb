```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m_cur=b_cur=0
    iter=10000
    n=len(x)
    l_rate=0.08
    for i in range(iter):
        y_pred=m_cur*x+b_cur
        cost=(1/n)*sum([ i**2 for i in (y-y_pred)])
        md=(-2/n)*sum(x*(y-y_pred))
        bd=(-2/n)*sum(y-y_pred)
        m_cur=m_cur-l_rate*md
        b_cur=b_cur-l_rate*bd
        print(" m= {} , b= {} ,cost= {} , iter= {} ".format(m_cur,b_cur,cost,i))
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
